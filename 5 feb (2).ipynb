```python
class Car:
    def __init__(self, make, model, color):
        self.make = make
        self.model = model
        self.color = color
    
    def start_engine(self):
        print(f"{self.color} {self.make} {self.model}'s engine is started.")
    
    def stop_engine(self):
        print(f"{self.color} {self.make} {self.model}'s engine is stopped.")


car1 = Car("Toyota", "Camry", "Blue")
car2 = Car("Honda", "Civic", "Red")


car1.start_engine()
car2.stop_engine()
```

```python
An instance of a class is referred to as an object. 
It is a physical manifestation of the class's design, complete with distinctive data (attribute values) and behavior (method executions). 
Car class objects car1 and car2 are shown in the example above. 
They can carry out the methods listed in the class and have unique characteristics (make, model, and color) (start and stop the engine).
```
